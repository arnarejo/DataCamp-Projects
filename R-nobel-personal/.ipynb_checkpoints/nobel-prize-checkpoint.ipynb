## Kaggle - Nobel Laureates, 1901-Present

This is recreation of Nobel Prize project on datacamp site. As a side project I have downloaded original data from Kaggle website and tried to recreate the same kernel. My primary target is to complete all challenges set in Datacamp challenge.

### 1. Load the Data

```r
# Loading in required libraries
library(tidyverse)
library(lubridate)
```

```r
# check the files in directory
list.files()
```

```r
# Reading in the Nobel Prize data
nobel <- read.csv("data/nobel.csv")
str(nobel)
```

```r
colnames(nobel)
```

### 2. Check total number of Nobel Prize given

```r
# count total number of nobel prizes given between 1901 and 2016
nobel %>%
    count()
```

### 3. Counting the prize by Gender

```r
# Counting the number of prizes won by male and female recipients.
nobel %>%
    group_by(Sex) %>%
    count()
```

### 4. Count prize count by country and display top 10 countries

```r
# Counting the number of prizes won by different nationalities and display top 10
nobel %>%
    group_by(Birth.Country) %>%
    count() %>%
    arrange(desc(n)) %>%
    head(10)
```

### 5. Calculate proportion of USA born winners for each decade

```r
# Calculating the proportion of USA born winners for each decade
proportion_usa_winners <- nobel %>%
    mutate(usa_born_winner = Birth.Country == "United States of America") %>%
    mutate(decade = floor(Year / 10) * 10) %>%
    group_by(decade) %>%
    summarize(proportion = mean(usa_born_winner, na.rm = TRUE))
proportion_usa_winners
```

### 6. Plotting the USA winners proportion for each decade

```r
# Plotting USA born winners
options(repr.plot.width=7, repr.plot.height=4)
ggplot(proportion_usa_winners, aes(x = decade, y = proportion)) +
    geom_line() + 
    geom_point() +
    scale_y_continuous(limits = c(0,1), labels = scales::percent_format())
```

### 7. Proportional of female winners by each category

```r
# Calculating the proportion of female winners for each decade
proportion_female_winners <- nobel %>%
    mutate(female_winner = Sex == "Female") %>%
    mutate(decade = floor(Year / 10) * 10) %>%
    group_by(decade, Category) %>%
    summarize(proportion = mean(female_winner, na.rm = TRUE))

# Plotting USA born winners
options(repr.plot.width=7, repr.plot.height=4)
ggplot(proportion_female_winners, aes(x = decade, y = proportion, color=Category)) +
    geom_line() + 
    geom_point() + 
    scale_y_continuous(limits = c(0,1), labels = scales::percent_format())
```

### 8. The first woman to win the Nobel Prize

```r
nobel %>%
    filter(Sex == "Female") %>%
    arrange(Year) %>%
    head(1)
```

### 9. Nobel laureates with more than 1 prize

```r
nobel %>%
    group_by(Full.Name) %>%
    count() %>%
    filter(n > 1) %>%
    arrange(desc(n))
```

### 10. Age when Nobel Prize Awarded

```r
nobel$Birth.Date <- as.Date(nobel$Birth.Date)
nobel_age <- nobel %>%
    mutate(Age = Year - year(Birth.Date))

ggplot(nobel_age, aes(x = Year, y = Age)) + geom_point() + geom_smooth()
```

### 11. Plot age difference by Categories

```r
ggplot(nobel_age, aes(x = Year, y = Age)) + geom_point() + geom_smooth() + facet_wrap(~Category)
```

### 12. Find oldest and youngest winners

```r
# Oldest
nobel_age %>%
    top_n(1, Age )

# Youngest
nobel_age %>%
    top_n(1, desc(Age))
```
